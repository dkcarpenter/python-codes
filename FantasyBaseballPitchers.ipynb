Import the csv file containing 6 separate sheets(for each year) of copied data from fangraphs, create a dictionary of dataframes(where each sheet is a dataframe)

```python
#Import csv files (minimum 70 IP)
import pandas as pd

# Specify the path to your Excel file
excel_file_path = 'C:/Users/dkcar/Downloads/MajorLeaguePitchersAdvancedStats.xlsx'

# Use pandas to read the Excel file
xls = pd.ExcelFile(excel_file_path)

# Create a dictionary of DataFrames, where keys are sheet names and values are DataFrames
dfs = {sheet_name: xls.parse(sheet_name) for sheet_name in xls.sheet_names}

# Access individual DataFrames using sheet names

__2023Standard = dfs[xls.sheet_names[0]]
__2023Advanced = dfs[xls.sheet_names[1]]
__2022Standard = dfs[xls.sheet_names[2]]
__2022Advanced = dfs[xls.sheet_names[3]]
__2021Standard = dfs[xls.sheet_names[4]]
__2021Advanced = dfs[xls.sheet_names[5]]
```

```python
__2023Advanced.head()
```

Merge corresponding dataframes together

```python
__2023Pitchers = pd.merge(__2023Standard,__2023Advanced,how = 'inner',on = 'Name')
__2022Pitchers = pd.merge(__2022Standard,__2022Advanced,how = 'inner',on = 'Name')
__2021Pitchers = pd.merge(__2021Standard,__2021Advanced,how = 'inner',on = 'Name')
```

Add TotalPoints, AveragePoints

```python
__2021Pitchers.columns

#Calculate total fantasy points and add to dataframes
__2021Pitchers['TotalPoints'] = 3*__2021Pitchers['IP'] + 5*__2021Pitchers['W'] - 3*__2021Pitchers['L'] - __2021Pitchers['H'] - __2021Pitchers['BB'] - __2021Pitchers['HBP'] - __2021Pitchers['IBB'] + __2021Pitchers['SO'] + 5*__2021Pitchers['SV'] + 2*__2021Pitchers['HLD'] - 2*__2021Pitchers['BS'] - 2*__2021Pitchers['ER']
__2022Pitchers['TotalPoints'] = 3*__2022Pitchers['IP'] + 5*__2022Pitchers['W'] - 3*__2022Pitchers['L'] - __2022Pitchers['H'] - __2022Pitchers['BB'] - __2022Pitchers['HBP'] - __2022Pitchers['IBB'] + __2022Pitchers['SO'] + 5*__2022Pitchers['SV'] + 2*__2022Pitchers['HLD'] - 2*__2022Pitchers['BS'] - 2*__2022Pitchers['ER']
__2023Pitchers['TotalPoints'] = 3*__2023Pitchers['IP'] + 5*__2023Pitchers['W'] - 3*__2023Pitchers['L'] - __2023Pitchers['H'] - __2023Pitchers['BB'] - __2023Pitchers['HBP'] - __2023Pitchers['IBB'] + __2023Pitchers['SO'] + 5*__2023Pitchers['SV'] + 2*__2023Pitchers['HLD'] - 2*__2023Pitchers['BS'] - 2*__2023Pitchers['ER']

#Calculate average fantasy points and add to dataframes
__2021Pitchers['AveragePoints'] = __2021Pitchers['TotalPoints'] / __2021Pitchers['G']
__2022Pitchers['AveragePoints'] = __2022Pitchers['TotalPoints'] / __2022Pitchers['G']
__2023Pitchers['AveragePoints'] = __2023Pitchers['TotalPoints'] / __2023Pitchers['G']
```

```python
__2023Pitchers.columns
```

Drop a few columns

```python
__2021Pitchers.drop(['Team_y'], axis=1, inplace=True)
__2022Pitchers.drop(['Team_y'], axis=1, inplace=True)
__2023Pitchers.drop(['Team_y'], axis=1, inplace=True)
```

Concatenate together

```python
concatenated_df = pd.concat([__2021Pitchers,__2022Pitchers,__2023Pitchers], axis=0, ignore_index=True)
```

```python
#Import fangraph triple A data (Min 70 IP)

#International League Standard:
df11 = pd.read_csv('fangraphs-minor-league-leaders (4).csv')
#International League Advanced:
df12 = pd.read_csv('fangraphs-minor-league-leaders (5).csv')

#Pacific League Standard
df13 = pd.read_csv('fangraphs-minor-league-leaders (6).csv')
#Pacific League Advanced
df14 = pd.read_csv('fangraphs-minor-league-leaders (7).csv')
```

```python
df11.columns
```

Drop Unneccessary Columns

```python
#df11.drop(['CG','ShO','TBF','R','WP','BK'], axis=1, inplace=True)
df12.drop(['Name','Team','Level','Age','ERA','IP'], axis=1, inplace=True)
#df13.drop(['CG','ShO','TBF','R','WP','BK'], axis=1, inplace=True)
df14.drop(['Name','Team','Level','Age','ERA','IP'], axis=1, inplace=True)
```

Merge and concatenate Data Frames

```python
df15 = pd.merge(df11,df12,how = 'inner',on = 'PlayerId')
df16 = pd.merge(df13,df14,how = 'inner',on = 'PlayerId')
concatdf = pd.concat([df15,df16],axis=0, ignore_index=True)
```

Add total points, Average Points columns

```python
concatdf['TotalPoints'] = 3*concatdf['IP'] + 5*concatdf['W'] - 3*concatdf['L'] - concatdf['H'] - concatdf['BB'] - concatdf['HBP'] - concatdf['IBB'] + concatdf['SO'] + 5*concatdf['SV'] - 2*concatdf['BS'] - 2*concatdf['ER']
concatdf['AveragePoints'] = concatdf['TotalPoints'] / concatdf['G']
```

Write to csv

```python
concatdf.to_csv('MinorLeaguePitchers.csv', index=False)
```
